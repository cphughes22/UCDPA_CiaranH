```python
# Showing how Higher-Level Art H1 scores have greatly increased since 2015

df_art = df.iloc[[23, 88, 152, 227, 306, 381, 458, 534]]
df_art
```

```python
# Comparing number of students completing higher and ordinary level Irish
# Irish results dataframe
# Higher-Level

df_irishhigh = df.iloc[[0, 64, 128, 203, 282, 357, 434, 511], [1, 2, 3, 21]]
df_irishhigh
```

```python
# Irish results dataframe
# Ordinary-Level

df_irishord = df.iloc[[2, 66, 130, 205, 284, 359, 436, 513], [1, 2, 3, 21]]
df_irishord
```

```python
# Plotting both Dataframes on a line graph. Higher-Level v Ordinary-Level

plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

df1 = df_irishhigh

df2 = df_irishord

ax = df1.plot(x='Year', y='num_total')
df2.plot(ax=ax, x='Year', y='num_total')
location = 0 # For the best location
legend_drawn_flag = True
plt.legend(["Higher-Level", "Ordinary-Level"], loc=0, frameon=legend_drawn_flag)

plt.xlabel('Year')
plt.ylabel('Students')
plt.title('Irish: Students Doing Higher and Ordinary Level')

plt.show()
```
